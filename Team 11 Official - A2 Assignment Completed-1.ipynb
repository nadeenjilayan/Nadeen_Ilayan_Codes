### Introduction to Model Selection and Evaluation

**Understanding the AIC Metric**

The Akaike Information Criterion (AIC) serves as a fundamental measure for model selection, particularly in time series analysis. It evaluates the goodness of fit of a model while penalizing the inclusion of additional parameters, thus preventing overfitting. Models with lower AIC values are preferred as they achieve a better balance between model complexity and fit to the data.

```python
# mean imputer
mean_imputer = SimpleImputer(strategy='mean')

# remove outliers to improve the model
def remove_outliers_from_columns(df, columns, threshold=1.5):
    # Select the specified columns
    selected_df = df[columns]
    
    # Calculate the IQR
    Q1 = selected_df.quantile(0.25)
    Q3 = selected_df.quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate the outlier bounds
    lower_bound = Q1 - threshold * IQR
    upper_bound = Q3 + threshold * IQR
    
    # Replace outliers with NaN
    outlier_removed_df = df.copy()  # Make a copy of the original DataFrame without modifying it
    for col in columns:
        outlier_removed_df.loc[~((outlier_removed_df[col] >= lower_bound[col]) & (outlier_removed_df[col] <= upper_bound[col])), col] = np.nan
    
    return outlier_removed_df

# Remove outliers from the specified columns
selected_numerical_columns = ['TotalProdMonths', 'TotalCompletionMonths',
       'ProducingDays', 'LiquidsProd_BBL', 'GasProd_MCF', 'CumLiquids_BBL',
       'CumGas_MCF']
Production_Data = remove_outliers_from_columns(Production_Data, selected_numerical_columns)

# apply mean into outliers
Production_Data[selected_numerical_columns] = mean_imputer.fit_transform(Production_Data[selected_numerical_columns])
```

```python
# Assuming you have a DataFrame 'Production_Data_Enhanced' with columns:
# 'Cluster', 'ProducingMonth', 'CumLiquids_BBL', and other features

# Define the target variable column
target_column = 'LiquidsProd_BBL'

# List to store model results
model_results = []

# Iterate over each cluster
for cluster in Production_Data_Enhanced['Cluster'].unique():
    # Filter data for the current cluster
    cluster_data = Production_Data_Enhanced[Production_Data_Enhanced['Cluster'] == cluster]
    
    # Define the columns to drop (adjust if needed)
    columns_to_drop = ['API_UWI', 'ProducingMonth', target_column, 'Cluster']
    
    # Create feature matrix X and target vector y
    X = cluster_data.drop(columns=columns_to_drop, axis=1)
    y = cluster_data[target_column]
    
    # One-hot encode categorical variables in X
    X_encoded = pd.get_dummies(X)
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
    
    # Initialize the linear regression model
    model = LinearRegression()
    
    # Fit the model to the training data
    model.fit(X_train, y_train)
    
    # Make predictions on the testing data
    y_pred = model.predict(X_test)
    
    # Evaluate the model (calculate Mean Squared Error)
    mse = mean_squared_error(y_test, y_pred)
    
    # Store the results
    model_results.append({'Cluster': cluster, 'MSE': mse})

    r2 = r2_score(y_test, y_pred)

    n = len(y_test)

    # Número de características (parámetros del modelo) + 1 (intercepto)
    k = X_train.shape[1] + 1

    residuals = y_test - model.predict(X_test)
    # Suma de los cuadrados de los residuos
    RSS = np.sum(residuals ** 2)

    # Logaritmo de la verosimilitud maximizada (asumiendo distribución normal de errores)
    log_likelihood = -0.5 * n * np.log(2 * np.pi) - 0.5 * n * np.log(RSS / n) - n / 2

    # Cálculo del AIC
    AIC = -2 * log_likelihood + 2 * k
        
    # Store the results
    model_results.append({'Cluster': cluster, 'MSE': mse, 'R2':r2, 'AIC':AIC})


# Convert the results to a DataFrame
model_results_df = pd.DataFrame(model_results)
model_results_df = model_results_df.dropna()

# Print or analyze the model results as needed
print(model_results_df)
```

```python
def forecast_cluster(df, name_cluster, num_cluster, type_prod):
   from statsmodels.tsa.statespace.sarimax import SARIMAX
   from pmdarima import auto_arima 

   df_arima = df.set_index('ProducingMonth')[[name_cluster,type_prod]]
   df_arima = df_arima[df_arima[name_cluster] == num_cluster]
   # Summing up production for each month if the data contains multiple wells or entries per month
   monthly_data = df_arima[type_prod].resample('M').mean()
   # Splitting the data into train and test sets
   split_point = int(len(monthly_data) * 0.8)
   train, test = monthly_data[:split_point], monthly_data[split_point:]

    
    # Fit the SARIMAX model
   model = SARIMAX(monthly_data, order=(1,1,1), seasonal_order=(1,1,1,12), trend='ct')
   model_fit = model.fit(disp=False)
   
   forecast = model_fit.forecast(steps=24)

   # Crear un rango de fechas para los siguientes 24 meses
   forecast_index = pd.date_range(start=monthly_data.index[-1], periods=25, freq='M')[1:]

   # Crear un DataFrame con los valores pronosticados
   forecast_df = pd.DataFrame({'Predictions': forecast.apply(lambda x: "{:,.2f}".format(x))}, index=forecast_index)

   # Plotting the results
   plt.figure(figsize=(18, 6))
   plt.plot(monthly_data.index, monthly_data, label=f'Historical Data - {name_cluster} {num_cluster}')
   plt.plot(forecast_index, forecast, label='Forecast')
   plt.legend()
   plt.grid(axis='both')
   plt.show()
```
